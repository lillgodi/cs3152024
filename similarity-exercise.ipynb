```python
import pandas as pd
import numpy as np
from numpy.linalg import norm
```

```python
sentences = ["she wore a dress and red earrings",
            "the dress has a red wine stain",
            "tomorrow I will wear my new red dress",
            "the red dress in the photo resembles the red dress she is wearing",
            "short dress",
            "red lipstick"]

query = "red dress"

sentencesAndQuery = list(sentences)
sentencesAndQuery.append(query)
```

```python
wordLists = [x.split() for x in sentencesAndQuery]
allWords = [i for row in wordLists for i in row]
len(allWords)
```

```python
def createColForD(sentences,allWords):
    inSentence = {"d":['d1','d2','d3','d4','d5','d6','q']}
    
    for w in allWords:
        w_val = 0
        inSentence[w] = []
        
        for s in sentences:
            w_val = s.count(w)
            #print(w_val)
            inSentence[w].append(w_val)
                
    return inSentence
```

```python
df = pd.DataFrame(createColForD(sentencesAndQuery,allWords)).set_index(['d'])
df
```

```python
def cosineSimilarity(vec1,vec2):
    V1 = np.array(vec1)
    V2 = np.array(vec2)
    print(norm(V1))
    cosine = np.dot(V1,V2)/(norm(V1)*norm(V2))
    return cosine
```

```python
queryVec = df.loc['q'].to_list()
tups = []

for i,s in enumerate(sentences):
    df_index = i+1
    sVec = df.loc[f"d{df_index}"].to_list()
    tups.append((cosineSimilarity(queryVec,sVec),s))

sorted_tuple = sorted(tups, key=lambda x: x[0],reverse=True)
sorted_tuple
```
