```python
import pandas as pd
import numpy as np
from numpy.linalg import norm
```

```python
sentences = ["she wore a dress and red earrings",
            "the dress has a red wine stain",
            "tomorrow I will wear my new red dress",
            "the red dress in the photo resembles the red dress she is wearing",
            "short dress",
            "red lipstick"]

query = "red dress"

sentencesAndQuery = sentences 
sentencesAndQuery.append(query)
```

```python
wordLists = [x.split() for x in sentencesAndQuery]
allWords = [i for row in wordLists for i in row]
len(allWords)
```

```python
def createColForD(sentences,allWords):
    inSentence = {"d":['d1','d2','d3','d4','d5','d6','q']}
    
    for w in allWords:
        inSentence[w] = []
        
        for s in sentences:
            if w in s:
                inSentence[w].append(1)
            else:
                inSentence[w].append(0)
                
    return inSentence
```

```python
createColForD(sentencesAndQuery,allWords)
```

```python
df = pd.DataFrame(createColForD(sentences,allWords)).set_index(['d'])
df
```

```python
def cosineSimilarity(vec1,vec2):
    V1 = np.array(vec1)
    V2 = np.array(vec2)
    cosine = np.dot(V1,V2)/(norm(V1)*norm(V2))
    return cosine
```

```python
queryVec = df.loc['q'].to_list()
for i,s in enumerate(sentences):
    df_index = i+1
    print(df_index)
    sVec = df.loc[f"d{df_index}"].to_list()
    print(f"{cosineSimilarity(queryVec,sVec)}: {s}")
```
